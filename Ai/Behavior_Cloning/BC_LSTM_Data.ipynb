```python
full_df = pd.read_csv(r"C:\Users\SlayerDz\PycharmProjects\clash-royale-rl-agent\Ai\Behavior_Cloning\full_cleaned_dataset.csv")
full_df["action"].unique()
```

```python
action_mapping = {
    "wait" : 0,
    "mini pekka" : 1,
    "knight": 2,
    "arrows": 3,
    "goblins": 4,
    "giant": 5,
    "spear goblins": 6,
    "archers": 7,
    "fireball" : 8,
    "minions": 9,
    "musketeer": 10,
    "three musketeers" : 0,
    "goblin cage" : 11
}
full_df["action"] = full_df["action"].map(action_mapping)
full_df["action"]
```

```python
#splitting the data set into training and testing sets based on match id
train_df = full_df[(full_df["match_id"] <= 27) | ((full_df["match_id"] >= 1001) & (full_df["match_id"] <= 1017))]
test_df = full_df[(full_df["match_id"] > 27) & ((full_df["match_id"] < 1001) | (full_df["match_id"] > 1017))]
train_df.to_csv("train_cleaned_dataset.csv", index=False)
test_df.to_csv("test_cleaned_dataset.csv", index=False)
```
